```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# sklearn imports
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
# 0) Ensure we have a dataframe variable (data or df)
if 'data' in globals():
    df = data.copy()
elif 'df' in globals():
    df = df.copy()
else:
    raise NameError("No dataframe named 'df' or 'data' found. Load the KDD file into variable named df or data.")
```

```python
# 1) Assign NSL-KDD column names if missing (common cause of many KeyErrors)
kdd_feature_names = [
    "duration","protocol_type","service","flag","src_bytes","dst_bytes","land","wrong_fragment",
    "urgent","hot","num_failed_logins","logged_in","num_compromised","root_shell","su_attempted",
    "num_root","num_file_creations","num_shells","num_access_files","num_outbound_cmds",
    "is_host_login","is_guest_login","count","srv_count","serror_rate","srv_serror_rate",
    "rerror_rate","srv_rerror_rate","same_srv_rate","diff_srv_rate","srv_diff_host_rate",
    "dst_host_count","dst_host_srv_count","dst_host_same_srv_rate","dst_host_diff_srv_rate",
    "dst_host_same_src_port_rate","dst_host_srv_diff_host_rate","dst_host_serror_rate",
    "dst_host_srv_serror_rate","dst_host_rerror_rate","dst_host_srv_rerror_rate"
]
expected_cols = kdd_feature_names + ["label", "difficulty"]

# If columns are integer-coded (header=None), assign names
if list(df.columns)[:5] == list(range(min(5, len(df.columns)))):
    if len(df.columns) == len(expected_cols):
        df.columns = expected_cols
        print("Assigned NSL-KDD column names.")
    else:
        print("Warning: number of columns does not match expected KDD count. Provided columns:", len(df.columns))
else:
    # If label column missing but an equivalent exists, try to detect label-like columns
    if 'label' not in df.columns and any(col in df.columns for col in ['attack', 'class', 41]):
        print("Note: 'label' column not found but other columns exist. Please ensure the dataset has a label column named 'label'.")
```

```python
# 2) Create binary target  and attack_category (DoS/Probe/R2L/U2R) safely
if 'label' not in df.columns:
    raise KeyError("Column 'label' not found after assigning names. Check your input file or column order.")

if 'label_binary' not in df.columns:
    df['label_binary'] = df['label'].apply(lambda x: 0 if str(x).strip() == 'normal' else 1)
dos = {"back","land","neptune","pod","smurf","teardrop","apache2","udpstorm","processtable"}
probe = {"satan","ipsweep","nmap","portsweep","mscan","saint"}
r2l = {"ftp_write","guess_passwd","imap","multihop","phf","spy","warezclient","warezmaster","sendmail"}
u2r = {"buffer_overflow","loadmodule","perl","rootkit","ps","sqlattack","xterm"}

def map_category(lbl):
    l = str(lbl).lower()
    if l == 'normal':
        return 'normal'
    if l in dos:
        return 'DoS'
    if l in probe:
        return 'Probe'
    if l in r2l:
        return 'R2L'
    if l in u2r:
        return 'U2R'
    if any(s in l for s in ['dos','smurf','neptune','back','teardrop','pod']):
        return 'DoS'
    if any(s in l for s in ['scan','sweep','nmap','ipsweep','portsweep','probe']):
        return 'Probe'
    if any(s in l for s in ['ftp','guess','warez','imap','sendmail','phf']):
        return 'R2L'
    if any(s in l for s in ['buffer','root','loadmodule','perl']):
        return 'U2R'
    return 'Other'

if 'attack_category' not in df.columns:
    df['attack_category'] = df['label'].apply(map_category)

print("Target counts (binary):\n", df['label_binary'].value_counts())
print("Attack category counts:\n", df['attack_category'].value_counts().head(10))
```

```python
# 3) Build preprocessing pipeline
# Define categorical and numeric features
all_cols = [c for c in df.columns]
exclude = ['label', 'label_binary', 'attack_category', 'difficulty']
feature_cols = [c for c in all_cols if c not in exclude]
categorical_features = [c for c in ['protocol_type','service','flag'] if c in feature_cols]
numeric_features = [c for c in feature_cols if c not in categorical_features]

print("Numeric feature count:", len(numeric_features), "Categorical feature count:", len(categorical_features))

# Numeric pipeline: median impute + StandardScaler
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

# Categorical pipeline: constant impute + OneHotEncoder
onehot_kwargs = {'handle_unknown': 'ignore'}
try:
    onehot = OneHotEncoder(sparse_output=False, **onehot_kwargs)
except TypeError:
    onehot = OneHotEncoder(sparse=False, **onehot_kwargs)

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', onehot)
])

preprocessor = ColumnTransformer(transformers=[
    ('num', numeric_transformer, numeric_features),
    ('cat', categorical_transformer, categorical_features)
], remainder='drop', sparse_threshold=0)
```

```python
# 4) Split and Train
X = df[feature_cols]
y = df['label_binary'].astype(int)
if len(y.unique()) < 2:
    raise ValueError("Target y contains only one class (all normal or all attack). Check your label column.")

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)
print("Train shape:", X_train.shape, "Test shape:", X_test.shape)
rf_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(n_estimators=200, random_state=42, n_jobs=-1))
])

# Fit
rf_pipeline.fit(X_train, y_train)
```

```python
# Predict + evaluate
y_pred = rf_pipeline.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred, target_names=['normal','attack']))
print("Confusion matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
import pandas as pd
from datetime import datetime

def generate_alerts(model, X_test, threshold=0.5):
    alerts = []

    probs = model.predict_proba(X_test)[:, 1]

    for i, prob in enumerate(probs):
        if prob >= threshold:
            alerts.append({
                "Time": datetime.now(),
                "Row_ID": i,
                "Attack_Probability": prob,
                "Alert": "INTRUSION DETECTED"
            })

    return pd.DataFrame(alerts)
```

```python
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, accuracy_score

# -----------------------------
# Page settings
# -----------------------------
st.set_page_config(page_title="SentinelNet Dashboard", layout="wide")

st.title("🛡 SentinelNet – Network Intrusion Detection Dashboard")

# -----------------------------
# Load data
# -----------------------------
@st.cache_data
def load_data():
    return pd.read_csv("KDDTrain+.txt", header=None)

data = load_data()

# Column names (important)
columns = [
    "duration","protocol_type","service","flag","src_bytes","dst_bytes","land",
    "wrong_fragment","urgent","hot","num_failed_logins","logged_in","num_compromised",
    "root_shell","su_attempted","num_root","num_file_creations","num_shells",
    "num_access_files","num_outbound_cmds","is_host_login","is_guest_login",
    "count","srv_count","serror_rate","srv_serror_rate","rerror_rate",
    "srv_rerror_rate","same_srv_rate","diff_srv_rate","srv_diff_host_rate",
    "dst_host_count","dst_host_srv_count","dst_host_same_srv_rate",
    "dst_host_diff_srv_rate","dst_host_same_src_port_rate",
    "dst_host_srv_diff_host_rate","dst_host_serror_rate",
    "dst_host_srv_serror_rate","dst_host_rerror_rate","dst_host_srv_rerror_rate",
    "label","difficulty"
]

data.columns = columns

# Binary label
data["binary_label"] = data["label"].apply(lambda x: "Normal" if x=="normal" else "Attack")

# -----------------------------
# Sidebar
# -----------------------------
st.sidebar.header("Navigation")
option = st.sidebar.selectbox(
    "Select View",
    ["Dataset Overview", "Traffic Distribution", "Attack Analysis", "Alerts"]
)

# -----------------------------
# Dataset Overview
# -----------------------------
if option == "Dataset Overview":
    st.subheader("📄 Dataset Overview")
    st.write("Shape of Dataset:", data.shape)
    st.dataframe(data.head())

# -----------------------------
# Traffic Distribution
# -----------------------------
elif option == "Traffic Distribution":
    st.subheader("📊 Traffic Distribution")

    col1, col2 = st.columns(2)

    with col1:
        st.write("Normal vs Attack")
        fig, ax = plt.subplots()
        sns.countplot(x=data["binary_label"], ax=ax)
        st.pyplot(fig)

    with col2:
        st.write("Protocol Distribution")
        fig, ax = plt.subplots()
        sns.countplot(y=data["protocol_type"], ax=ax)
        st.pyplot(fig)

# -----------------------------
# Attack Analysis
# -----------------------------
elif option == "Attack Analysis":
    st.subheader("🚨 Attack Analysis")

    fig, ax = plt.subplots(figsize=(10,4))
    data[data["label"]!="normal"]["label"].value_counts().plot(kind="bar", ax=ax)
    plt.title("Attack Type Distribution")
    st.pyplot(fig)

# -----------------------------
# Alerts (Simulation)
# -----------------------------
elif option == "Alerts":
    st.subheader("🔔 Alert Log (Simulated)")

    alerts = data[data["binary_label"]=="Attack"].sample(20)
    st.dataframe(alerts[["protocol_type","service","flag","src_bytes","dst_bytes","label"]])

    st.success("⚠ Intrusion Alerts Generated Successfully")

# -----------------------------
# Footer
# -----------------------------
st.markdown("---")
st.markdown("**SentinelNet | AI-Powered Network Intrusion Detection System**")
```

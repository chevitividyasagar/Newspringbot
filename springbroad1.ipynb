```python
import zipfile
import os
import pandas as pd

# Correct path
zip_path = "/mnt/data/KDDTrain+.txt.zip"
extract_dir = "/mnt/data/kdd_extracted"

# Check if file exists
print("Exists?", os.path.exists(zip_path))

# Extract ZIP
with zipfile.ZipFile(zip_path, 'r') as z:
    z.extractall(extract_dir)

# Load the extracted txt file
txt_file = os.path.join(extract_dir, "KDDTrain+.txt")
df = pd.read_csv(txt_file, header=None)

df.head()
```
